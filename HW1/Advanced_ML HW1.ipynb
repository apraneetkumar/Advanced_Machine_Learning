### Question 4 | Part 1

1) R Squared : It assumes that all the independent variables used are going to impact the prediction of the dependent variable. Also, with every incremental independent variable added to the model, R-squared always increases

2) Adjusted R Squared : This doesn't consider all the independent variables to affect the result of the model but only the ones that impact the model performance. Adjusted R Squared value only increases if the variable actually affects the dependent variable prediction.

Also, in Adjusted R Squared there's a penalty for adding more features which can be inferred from the formula below -

Adjusted R squared = 1 - (1-R Squared)*(n-1)/(n-p-1)

### Question 4 | Part 2

Yes, overfitting is possible in Linear regression. The model can overfit in the following scenarios -

1) The model can overfit when the number of dependent variables is large or comparable to the number of data points available. Such datasets make a prediction that may be perfect in training data as it will fit for for every data point but such model will not be generalized leading to a high error in Test Data.

2) If we fit a model with independent variables that are not related to the dependent variable then the model will try to fit over noise which may lead to a low error in training data but the model performance will be poos in test data as the features are not related to the dependent variable.

3) If we try to fit a model where the dependent variable's form is that of a higher-degree polynomial, then the regression line will be a poor fit as it cannot account for the variability of the dependent variable.

We can try to tackle overfitting by using the below methods -

1) By using regularization techniques such as Ridge, Lasso, or Elastic Net we can reduce the coefficients of some of the independent variables to zero, which will lead to a more generalized model.

2) Use sklearn.linear_model.Lasso and sklearn.linear_model.Ridge classes to do a 5-fold cross validation using sklearn's KFold. For the sweep of the regularization parameter, we will look at a grid of values ranging from  =1010  to  =106 .

In Python, you can consider this range of values as follows: alpha = 10**numpy.linspace(6,-6,100) so that you can generate 100 uniform values between -6 to 6 as power series.

Fit the 2 regression models with scaled data and report the best chosen  based on cross validation as well as the corresponding scoring metric. The cross validation should happen on your training data using MSE as the scoring metric. (5 pts)

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold

alphas = 10**np.linspace(10,-6,100) 

kf = KFold(n_splits=5)

mse_lasso=[]

mse_ridge=[]

#Ridge Regularization: cross validation using 5 folds to find alpha
for alpha in alphas:
    
    ridge = Ridge(alpha = alpha)
    
    mse_temp=[]
    
    for train_index, test_index in kf.split(X_train_scaled):
        X_traincv, X_testcv = X_train_scaled.iloc[train_index], X_train_scaled.iloc[test_index]
        Y_traincv, Y_testcv = Y_train.iloc[train_index], Y_train.iloc[test_index]
        ridge_fit = ridge.fit(X_traincv, Y_traincv)
        Y_pred = ridge_fit.predict(X_testcv)
        mse = mean_squared_error(Y_testcv, Y_pred)
        mse_temp.append(mse)
    
    mse_ridge.append(np.mean(mse_temp))

#best alpha for ridge based on cross validation and using MSE as the evaluation metric

best_ridge = alphas[mse_ridge.index(min(mse_ridge))]

min_mse_ridge= min(mse_ridge)

print("The best value of alpha for ridge is: "+ str(round(best_ridge,2)),"and lowest mse_ridge is: "+ str (round(min_mse_ridge,2)) )

#Lasso Regularization: cross validation using 5 folds to find alpha
for alpha in alphas:
    
    lasso = Lasso(alpha = alpha)
    
    mse_temp=[]
    
    for train_index, test_index in kf.split(X_train_scaled):
        X_traincv, X_testcv = X_train_scaled.iloc[train_index], X_train_scaled.iloc[test_index]
        Y_traincv, Y_testcv = Y_train.iloc[train_index], Y_train.iloc[test_index]
        lasso_fit = lasso.fit(X_traincv, Y_traincv)
        Y_pred = lasso_fit.predict(X_testcv)
        mse = mean_squared_error(Y_testcv, Y_pred)
        mse_temp.append(mse)
    
    mse_lasso.append(np.mean(mse_temp))

#best alpha for lasso based on cross validation 
best_lasso = alphas[mse_lasso.index(min(mse_lasso))]

min_mse_lasso= min(mse_lasso)

print("The best value of alpha for lasso is: "+ str(round(best_lasso,2)),"and lowest mse_lasso is: "+str(round(min_mse_lasso,2)))
```

7) Why did we have to scale the data in ridge and lasso regression? (3 pts)

Both Lasso and Ridge regularization methods use the coefficients to calculate the L1 norm and L2 norm respectively, which in turn are used to calculate penalties.

L1 Norm (for Lasso) is the sum of absolute values of the coefficients (beta values) and L2 Norm (For Ridge) is the sum of squares of coefficients (beta values). Thus, the penalty will inadvertently be highly dependent on the coefficients of the independent variables that have a higher scale. Hence, the variables are scaled as post standardization, the variables will have unit variance (range of variables will be -1 to 1) and the mean will be 0.
